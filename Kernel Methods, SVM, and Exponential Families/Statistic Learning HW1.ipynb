# 1. Kernel Methods

## Problem 1

### 1   
已知$k(x,y) = 1 + x^Ty, \ \forall \ \mathit{X} = \mathbb{R}^d, \ \mathit{F} = \mathbb{R}^{d+1}$，   
根据$dot \ product \ of \ polynomials和kernel \ function，\\ φ(x) = (1,x), \ φ(y) = (1,y), \ k(x,y) = (1,x)\left(\begin{matrix}1 \\y\end{matrix}\right) = 1+x^Ty \\ \Rightarrow 若让d+1，则φ(x) = (1,\sqrt{2}x,x), \ φ(y) = (1,\sqrt{2}y,y), \\ k(x,y) = (1,\sqrt{2}x,x)\left(\begin{matrix}1 \\ \sqrt{2}y\\y\end{matrix}\right) = (1+x^Ty)^2 \\ \Rightarrow 可以推出，For \ General: k(x,y)=(1+xy)^n, \ \forall \ \mathit{X} = \mathbb{R}$

### 2
从理论来看，$k(x,y) = ⟨φ(x), φ(y)⟩$，故会是两个同样$function的matrix$做内积。   
欲使$k(x,y) = xy - 1 = -1 + xy，则可令φ(x) = (i,x), \ φ(y) = (i,y) \\ \Rightarrow k(x,y) = (i,x)\left(\begin{matrix}i \\y\end{matrix}\right) = -1+xy$   
然而上述推导与$k(x,y) = xy - 1, \ \forall \ \mathit{X} = \mathbb{R}$相矛盾，因为$i$是虚数，故$k(x,y) = xy - 1, \ \forall \ \mathit{X} = \mathbb{R}$不是$kernel function$

### 3
下证$k(x, y) = min(x, y), \ \forall \ \mathit{X} = [0, 1]$是$kernel \ function$:    
欲证$min(x, y) = ⟨φ(x), φ(y)⟩$    
$min(x,y) = \int \mathbb{1}_{(0,x)}\mathbb{1}_{(0,y)} = ⟨\mathbb{1}_{(0,x)}, \mathbb{1}_{(0,y)}⟩ = ⟨φ(x), φ(y)⟩ \\ \Rightarrow k(x, y) = min(x, y), \ \forall \ \mathit{X} = [0, 1]是kernel \ function$

```python
import numpy as np
from sklearn.kernel_ridge import KernelRidge
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 150
```

```python
x = np.random.uniform(0,1,30) #x
e = np.random.normal(0, 1, 30) #noise
f = lambda x,e: ((-5) + 20*x - 16*(x**2) + 0.4*e)
y = f(x,e) #y
```

```python
dt = pd.DataFrame({'x':x,r'$\epsilon$':e,'y':y})

display(dt)
```

## 2

```python
X = x.reshape(30, 1)
#fit
clf_0001 = KernelRidge(alpha=0.001,kernel='rbf', gamma=0.1)
clf_0001.fit(X, y)

clf_001 = KernelRidge(alpha=0.01,kernel='rbf', gamma=0.1)
clf_001.fit(X, y)

clf_01 = KernelRidge(alpha=0.1,kernel='rbf', gamma=0.1)
clf_01.fit(X, y)

clf_1 = KernelRidge(alpha=1.0,kernel='rbf', gamma=0.1)
clf_1.fit(X, y)

clf_2 = KernelRidge(alpha=2.0,kernel='rbf', gamma=0.1)
clf_2.fit(X, y)

clf_3 = KernelRidge(alpha=3.0,kernel='rbf', gamma=0.1)
clf_3.fit(X, y)
```

```python
#(1+xy)^n n=1,2,9
def k1(X,y):
    return(1 + X.T.dot(y))

def k2(X,y):
    return((1 + X.T.dot(y))**2)

def k9(X,y):
    return((1 + X.T.dot(y))**9)

    
clf_01_k1 = KernelRidge(alpha=0.001,kernel = k1)
clf_01_k1.fit(X,y)

clf_01_k2 = KernelRidge(alpha=0.001,kernel = k2)
clf_01_k2.fit(X,y)

clf_01_k9 = KernelRidge(alpha=0.001,kernel = k9)
clf_01_k9.fit(X,y)
#min(x,y)
def kmin(X, y):
    return np.minimum(X.T, y)

clf_01_kmin = KernelRidge(alpha=0.001,kernel = kmin)
clf_01_kmin.fit(X,y)
#exp(-(x-y)^2)
def ke(X,y):
    return np.exp((X.T-y)**2)

ke(X,y).shape
clf_01_ke = KernelRidge(alpha=0.001,kernel = ke)
clf_01_ke.fit(X,y)
```

使用$kernel \ ridge \ regression$生成回归模型，以下设定$kernel = RBF$，并使用不同的$\lambda$值:    
     
    
可以看出随着$\lambda$值上升，模型越趋和$Target \  Function$不像

```python
t = lambda x: ((-5) + 20*x - 16*(x**2))
ty = [t(i) for i in x]
##lambda
dt = pd.DataFrame({'x':x,'y':y,'ty':ty,
                   'clf_0001':clf_0001.predict(X),
                   'clf_001':clf_001.predict(X),
                   'clf_01':clf_01.predict(X),
                   'clf_1':clf_1.predict(X),
                   'clf_2':clf_2.predict(X),
                   'clf_3':clf_3.predict(X)})

sns.set_theme()
sns.scatterplot(data=dt,x='x', y = 'y')
sns.lineplot(data=dt ,x='x', y = 'ty', color='tomato',linewidth=2.5, label='Target function')
sns.lineplot(data=dt ,x='x', y = 'clf_0001', color='tab:green',linewidth=1, label=r'$\lambda = 0.001$')
sns.lineplot(data=dt ,x='x', y = 'clf_001', color='orchid',linewidth=1, label=r'$\lambda = 0.01$')
sns.lineplot(data=dt ,x='x', y = 'clf_01', color='gold',linewidth=1, label=r'$\lambda = 0.1$')
sns.lineplot(data=dt ,x='x', y = 'clf_1', color='cornflowerblue',linewidth=1, label=r'$\lambda = 1$')
sns.lineplot(data=dt ,x='x', y = 'clf_2', color='sandybrown',linewidth=1, label=r'$\lambda = 2$')
sns.lineplot(data=dt ,x='x', y = 'clf_3', color='mediumaquamarine',linewidth=1, label=r'$\lambda = 3$')
plt.show()
```
